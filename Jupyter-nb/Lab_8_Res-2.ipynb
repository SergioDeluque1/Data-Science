### **Ejercicio #1: (0.8)**
Dados los siguientes vectores

\begin{equation*}
 \vec{u} = (12, 51, -1),
 \hspace{0.5cm}
 \vec{v} = (2, 11, 57),
\hspace{0.5cm}
 \vec{w} = (66, 9, 15),
 \hspace{0.5cm}
 \vec{z} = (1, 33, -8),
\end{equation*}

realizar las operaciones indicadas:

a). **(0.2)** $ (\vec{u} + \vec{z}- \vec{v}) \cdot (\vec{z}-3\vec{w} + \vec{v}).$

b). **(0.2)** $\| \vec{z}- \vec{w}\|- \|\vec{w} \times \vec{u}\|.$

c). **(0.2)** $\|\vec{u}\|(\vec{v} + \vec{z}) + \| 2\vec{z}- 3\vec{w}\|\vec{z}.$

d). **(0.2)** Hallar dos vectores $\vec{p}$ y $\vec{h}$ tales que $\vec{v} = \vec{p} + \vec{h}$ con $\vec{p}$ paralelo a $\vec{u}$ y $\vec{h}$ ortogonal a $\vec{u}$.

```python
# Solución Ejercicio 1. (No olvide imprimir las respuestas de sus cálculos)

# Importar la bibliotecas numpy con el alias np y math.
import numpy as np
import math

# Definir los vectores u,v,w,z

# Definición de los vectores
u = np.array([12, 51, -1])
v = np.array([2, 11, 57])
w = np.array([66, 9, 15])
z = np.array([1, 33, -8])

# (a):
pt1_a = u + z - v
pt2_a = z - 3 * w + v
res_1a = np.dot(pt1_a, pt2_a)

# (b):
dif_1b = z - w
norma_dif = math.sqrt(np.dot(dif_1b, dif_1b))
cruz = np.cross(w, u)
norma_cruz = math.sqrt(np.dot(cruz, cruz))
res_1b = norma_dif - norma_cruz

# (c):
norma_u = math.sqrt(np.dot(u, u))
suma_vz = v + z
pt1_c = norma_u * suma_vz
combinado = 2 * z - 3 * w
norma_combinado = math.sqrt(np.dot(combinado, combinado))
pt2_c = norma_combinado * z
res_c = pt1_c + pt2_c

# (d):
k = np.dot(u, v) / np.dot(u, u)
p = k * u
h = v - p
verificacion = np.dot(u, h)

# Imprimir las respuestas
print(f"a) Resultado: {res_1a:.2f}")
print(f"b) Resultado: {res_1b:.2f}")
print(f"c) Resultado: {res_c}")
print(f"d) p (paralelo a u): {p}")
print(f"   h (ortogonal a u): {h}")
print(f"   Verificación u·h = {verificacion:.2e} (debe ser ~0)")
```

### **Ejercicio #3: (0.6)**
Considere la matriz

\begin{equation*}
 A =
\begin{pmatrix}
 -k & k-1 & k+1 & k & 0\\
3 & 1 & 1 & 3 & -1\\
4-k & k & k+3 & -k & 1\\
-2 & 3 & 4 & 6 & 2\\
0 & 2 & 1 & 0 & 4
\end{pmatrix}.
\end{equation*} \


a). **(0.3)**  Halle los valores de $k$ para los cuales la matriz $A$ tiene inversa.

b). **(0.3)**  Halle los valores de $k$ para los cuales $\det(A) = 6+2k$.


```python
# Solución Ejercicio 3. (No olvide imprimir las respuestas de sus cálculos)

# (a):
import sympy as sp

k = sp.Symbol('k', real=True)


A = sp.Matrix([
    [-k,    k-1,  k+1,  k,  0],
    [3,     1,    1,    3, -1],
    [4-k,   k,    k+3, -k,  1],
    [-2,    3,    4,    6,  2],
    [0,     2,    1,    0,  4]
])


det_A = A.det()
print("Determinante de A:")
print(det_A)


k_no_inversa = sp.solve(det_A, k) # Valores de k para los cuales A no tiene inversa i.e det(A) = 0

print("\n A tiene inversa para todo k excepto:")
for val in k_no_inversa:
    print(f"  k ≠ {val}")

```

```python
# Solución Ejercicio 3. (No olvide imprimir las respuestas de sus cálculos)

# (b):
equation = sp.Eq(det_A, 6 + 2*k)
k_solutions = sp.solve(equation, k)

print("\n(b) Valores de k para los cuales det(A) = 6 + 2k:")
print(k_solutions)

```


### **Ejercicio #4: (0.7)**
Sean $A(x_1,y_1), \ B(x_2,y_2)$ y $C(x_3,y_3)$ tres puntos en el plano $xy$. El área del triángulo formado por dichos puntos $A,B,C$ es dada por la expresión  

\begin{equation*}
 \mbox{Área} = \frac{1}{2} \left | \det
\begin{pmatrix}
x_1 & y_1 & 1\\
x_2 & y_2 & 1 \\
x_3 & y_3 & 1
\end{pmatrix} \right|,
\end{equation*}

donde el símbolo $|.|$ denota el valor absoluto.

a). **(0.3)** Sean $A(-5,-2), \ B(-7,7)$ y $C(9,2)$. Halle el área del triángulo formado por los puntos $A, \ B, \ C$.  

b). **(0.4)**  ¿Para qué valores de $k$ el **cuadrado del área del triángulo** formado por los puntos $A(-4,k), \ B(1,k)$ y $C(k,3)$ es igual a $100\ \mbox{cm}^4$?

```python
# Solución Ejercicio 4. (No olvide imprimir las respuestas de sus cálculos)

# (a): 
import sympy as sp

x1, y1 = -5, -2
x2, y2 = -7, 7
x3, y3 = 9, 2


M = sp.Matrix([
    [x1, x2, x3],
    [y1, y2, y3],
    [ 1,  1,  1]
])


det_M = M.det()
print("Determinante del 3×3:", det_M)


area = sp.Abs(det_M) / 2
print("Área del triángulo:", area)


```

```python
# Solución Ejercicio 4. (No olvide imprimir las respuestas de sus cálculos)

# (b):

import sympy as sp

k = sp.symbols('k', real=True)


area_b_squared = 25*(k - 7)**2/4
eq = sp.Eq(area_b_squared, 100)

solutions_k = sp.solve(eq, k)

print("Valores de k para los cuales el cuadrado del área es 100:")
print(solutions_k)

```
